# **IMPLEMENTASI PCA PADA DATASET MEDIS BERDIMENSI TINGGI**
Dataset: Breast Cancer Wisconsin (Diagnostic)

**Tujuan**


1. Menerapkan Principal Component Analysis (PCA) pada dataset medis berdimensi tinggi.
2. Mengurangi dimensi fitur dari 30 menjadi beberapa komponen utama.
3. Menganalisis explained variance dari setiap komponen.
4. Menguji performa klasifikasi sebelum dan sesudah PCA.

# **Data Cleaning & Preprocessing**

```python
# Drop kolom tidak relevan
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
# Pisahkan fitur dan target
X = df.drop('diagnosis', axis=1)
y = df['diagnosis'].map({'M': 1, 'B': 0})

print("Jumlah fitur:", X.shape[1])
```

# **Standardisasi Data**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# **PCA**

```python
# PCA Tanpa Reduksi (Analisis Varians)
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_scaled)
```

```python
# Explained Variance Ratio
explained_variance = pca_full.explained_variance_ratio_

plt.figure(figsize=(10,5))
plt.plot(
    range(1, len(explained_variance) + 1),
    np.cumsum(explained_variance),
    marker='o'
)
plt.xlabel("Jumlah Principal Component")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance PCA")
plt.grid()
plt.show()
```

```python
# PCA dengan Reduksi Dimensi
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled)

print("Explained Variance (2 PC):", pca_2.explained_variance_ratio_)
```

```python
# Visualisasi PCA 2D
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=X_pca_2[:,0],
    y=X_pca_2[:,1],
    hue=y.map({1:'Malignant', 0:'Benign'}),
    palette=['red','blue'],
    alpha=0.7
)
plt.title("Visualisasi PCA (2 Komponen)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
# Klasifikasi TANPA PCA (Baseline)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model_no_pca = LogisticRegression(max_iter=1000)
model_no_pca.fit(X_train, y_train)

y_pred_no_pca = model_no_pca.predict(X_test)

print("Accuracy TANPA PCA:", accuracy_score(y_test, y_pred_no_pca))
print(classification_report(y_test, y_pred_no_pca))
```

```python
# Klasifikasi DENGAN PCA
pca_10 = PCA(n_components=10)
X_pca_10 = pca_10.fit_transform(X_scaled)

X_train, X_test, y_train, y_test = train_test_split(
    X_pca_10, y, test_size=0.2, random_state=42
)

model_pca = LogisticRegression(max_iter=1000)
model_pca.fit(X_train, y_train)

y_pred_pca = model_pca.predict(X_test)

print("Accuracy DENGAN PCA:", accuracy_score(y_test, y_pred_pca))
print(classification_report(y_test, y_pred_pca))
```

# **Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_pca)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (PCA)")
plt.show()
```
